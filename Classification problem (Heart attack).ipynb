```python
#Q-Q plot

import scipy.stats as stat
import pylab
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist(bins=20)
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist='norm',plot=pylab)
    plt.title(feature)
    plt.show()
    
for feature in cont_feature:
    plot_data(df,feature)
```

### Feature Engineering

#### Gaussian tranformation:

```python
df['oldpeak_sqrt']=(df['oldpeak'])**0.5

plot_data(df,'oldpeak_sqrt')
```

```python
df['thalachh_boxcox'],parameters=stat.boxcox(df['thalachh'])
print(parameters)
plot_data(df,'thalachh_boxcox')
```

```python
df['chol_boxcox'],parameters=stat.boxcox(df['chol'])
print(parameters)
plot_data(df,'chol_boxcox')
```

```python
df['trtbps_boxcox'],parameters=stat.boxcox(df['trtbps'])
print(parameters)
plot_data(df,'trtbps_boxcox')
```

```python
df.drop(['trtbps', 'chol', 'fbs','thalachh','oldpeak'],axis=1,inplace=True)
```

```python
# Feature Scalling: there are some outliers in datasets so i will use robust scaler here
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
```

```python
x=df.drop(['output'],axis=1)
y=df['output']
```

```python
scaler.fit(x)
```

```python
transform=scaler.transform(x)
```

```python
data=pd.concat([pd.DataFrame(transform,columns=x.columns),y],axis=1)
```
